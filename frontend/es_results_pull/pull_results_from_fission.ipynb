```python
import requests
import pandas as pd

BASE_URL = "http://localhost:9090/query-results"

def pull_index(index, limit=500000):
    url = f"{BASE_URL}?index={index}&limit={limit}"

    response = requests.get(url, timeout=300)
    response.raise_for_status()

    data = response.json()

    return pd.DataFrame(data)
```

```python
# Pull the main combined results index
df = pull_index("ml_results")

print(df.shape)

df.head()
```

```python
bitcoin_df = df[df["coin"] == "bitcoin"].copy()
ethereum_df = df[df["coin"] == "ethereum"].copy()
bnb_df = df[df["coin"] == "bnb"].copy()
xrp_df = df[df["coin"] == "xrp"].copy()
usdt_df = df[df["coin"] == "usdt"].copy()

spike_events_df = df.copy()
```

```python
# Pull the new NLP indices

ml_spike_events_df = pull_index("ml_spike_events", limit = 1000)

ml_daily_sentiment_df = pull_index("ml_daily_sentiment", limit = 1000)

ml_narratives_df = pull_index("ml_narratives", limit = 1000)

ml_correlations_df = pull_index("ml_correlations", limit = 1000)

print("ml_spike_events:", ml_spike_events_df.shape)
print("ml_daily_sentiment:", ml_daily_sentiment_df.shape)
print("ml_narratives:", ml_narratives_df.shape)
print("ml_correlations:", ml_correlations_df.shape)
```

```python
df.to_csv("ml_results_full_from_es.csv", index=False)

bitcoin_df.to_csv("bitcoin_results.csv", index=False)
ethereum_df.to_csv("ethereum_results.csv", index=False)
bnb_df.to_csv("bnb_results.csv", index=False)
xrp_df.to_csv("xrp_results.csv", index=False)
usdt_df.to_csv("usdt_results.csv", index=False)

spike_events_df.to_csv("spike_events.csv", index=False)

ml_spike_events_df.to_csv("ml_spike_events.csv", index=False)
ml_daily_sentiment_df.to_csv("ml_daily_sentiment.csv", index=False)
ml_narratives_df.to_csv("ml_narratives.csv", index=False)
ml_correlations_df.to_csv("ml_correlations.csv", index=False)
```
